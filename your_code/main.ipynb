#### Import all the libraries that are necessary.

#### Open the ` Employee.csv` file and name your dataset `employee`.

```python
employee=pd.read_csv('Employee.csv')
employee.head()
```

#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
# Method 1: Using groupby

employee['Department'].value_counts(dropna=False)
#Overview of the different departments (3 options only)

employee2=employee[['Department','Title', 'Salary']]
employee2=employee2.groupby(['Department', 'Title'], as_index=False)['Salary'].mean()
employee2
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python
# Method 2: using pivot table

employee3 = employee.pivot_table(index= ['Department','Title'], values = ['Salary'], aggfunc = {'Salary':['mean']}).reset_index()
employee3
```

#### We want to display the number of years of work experience and the salary for each Title. Add Years as a column in your pivot table.
*Hint: Use Pandas doc(see References in README.md) to explore the pivot_table() arguments*

```python

employee4 = employee.pivot_table(index= ['Department','Title'], values = ['Salary','Years'], aggfunc = ['mean']).reset_index()
employee4
```

#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
employee5 = employee.pivot_table(index= ['Department','Title'], values = ['Salary', 'Name'], aggfunc = {'Salary':'mean','Name':'count'}).reset_index()
employee5
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
employee6 = employee.pivot_table(index= ['Department','Title'], values = ['Salary'], aggfunc = {'Salary':['median','min','max','std']}).reset_index()
employee6
```

#### Based on your comments, fill the missing values with an appropriate value.

```python
# First I'm gonna check that the NaN values are not generated due to NaN values present in the table
employee['Salary'].value_counts(dropna=False)

#there's no NaN values and only 1 employee/salary for the values concerned by NaN
#the standard deviation should then be 0 since there won't be any variation possible for those category

employee6.columns = ['Department','Title', 'Salary_max', 'Salary_median', 'Salary_min','Salary_std']
employee6

employee6['Salary_std']= employee6['Salary_std'].fillna(0)
employee6['Salary_std']=employee6['Salary_std'].round(2)
employee6
```

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
# how much money spent = total of salary = sum()
employee8 = employee.pivot_table(index= ['Department'], values = ['Salary', 'Name'], aggfunc = {'Salary':'sum','Name':'count'}).reset_index()
employee8
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

```python
#total years = sum of the years of experience of each employee by department and title
employee9 = employee.pivot_table(index= ['Department','Title'], values = ['Salary','Years'], aggfunc = {'Salary':'mean','Years':'sum'}).reset_index()
employee9
```

### Bonus

#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
#step1: create a function which takes out the max value and calculate the mean of the remaining values in the data
def mean_aggregate(x):
    y= x.sum() - x.max()
    z= x.count() - 1
    return y/z

#step2: applying the function on Salary while grouping by department
employee10 = employee.pivot_table(index= ['Department'], values = ['Salary'], aggfunc = mean_aggregate).reset_index()
employee10
```
